```python
"""
1) Criação dos diretórios que serão usados no script
2) Download dos arquivos do repositório de desenvolvimento para uso no Colab
"""
import os
import requests
from random import random

data_dir = 'data/'
train_path = 'data/extended_quotes_funds.csv'
calculador_path = 'optimizer.py'

if not os.path.exists(data_dir):
  os.mkdir(data_dir)
if not os.path.exists(train_path):
  a = requests.get('https://raw.githubusercontent.com/felipedmz/fiagro-marko/main/input/extended_quotes_funds.csv')
  open(train_path, 'wb').write(a.content)
#
if not os.path.exists(calculador_path):
  c = requests.get(f'https://raw.githubusercontent.com/felipedmz/fiagro-marko/main/optimizer.py?cb={random()}')
  open(calculador_path, 'wb').write(c.content)
```
